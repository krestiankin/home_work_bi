```python
import pandas as pd
from sqlalchemy import create_engine
```

>3. Посчитайте количество побед каждого участника, принимая за победу 1-е место. Сохраните результаты во вторичном DataFrame.


```python
df3 = df[df['place'] == '1'] # вибірка тільки 1 місць

df3_1 = df3.groupby(['name'])[['place']].count().sort_values(by='place', ascending=False)  # групування по учасникам та підрахунок з сортуванням

# print(df3.place.count()) # перевірка на кількість перших місць до групування
# print(df3_1.place.sum()) # перевірка на кількість перших місць після групування

df3_1.head(10)
```

>4. Используйте DataFrame из упражнения 3, чтобы показать для каждой комбинации подразделения и пола, участника с наибольшим количеством побед среди участников,когда-либо участвовавших в этом дивизионе.

```python
df4 =  df3_1.merge(df, how='inner', on='name')[['division', 'sex', 'name', 'place_x']].reset_index(drop=True)
df4.loc[df4.groupby(['division', 'sex'])['place_x'].idxmax()].reset_index(drop=True)
```

>5. Подключитесь к базе данных employees и выполните с помощью pandas последний запрос из финального проекта по SQL. Убедитесь, что результаты загружены в pandas DataFrame.


```python
# підключення до бази даних за допомогою алхімії
engine = create_engine('mysql+pymysql://root:1@localhost:3306/employees')
```

```python
# закидуємо SQL запит зі степпроекту в змінну 
query = '''
WITH windowCTE AS (
	SELECT s.emp_no, de.dept_no, s.salary,
	   SUM((s.salary / 12)) OVER (PARTITION BY de.dept_no
							        ORDER BY s.salary
                                    ROWS BETWEEN unbounded preceding and current ROW
                                ) AS salary_sum_up
	FROM salaries AS s
    JOIN dept_emp AS de USING(emp_no)
    WHERE CURDATE() BETWEEN s.from_date AND s.to_date           
			AND CURDATE() BETWEEN de.from_date AND de.to_date 
	ORDER BY dept_no, salary)

SELECT w.dept_no, e.first_name, e.last_name, w.salary_sum_up
	FROM employees AS e
    JOIN windowCTE AS w USING(emp_no)
    WHERE w.salary_sum_up <= 500000;
'''
```

```python
# читаємо пандасом та дивимось результат запиту
df_sql = pd.read_sql(query, engine)
df_sql.head(20)
```
