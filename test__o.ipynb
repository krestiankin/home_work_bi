```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline

```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [6, 7, 8, 9]),
ax.legend()
iris = sns.load_dataset('iris')
iris.head()
iris.info()
iris.describe()
iris.isnull().sum()
iris.duplicated().sum()
sns.get_dataset_names()
plt.figure(figsize = (6,4)) # встановити розмір графіку 

```

```python
sns.countplot(x= iris.species, data = iris); # показати кіл-к данних

plt.xlabel('Class of iris') # встановити назву по осі x
plt.ylabel('Count') # встановити назву по осі y
plt.title('Iris spes') # встановити назву графіку
plt.show() # показати графік
```

```python
#  data = iris, hue = забарвлення точок від категор данних, diag_kind - тип графіка
sns.pairplot(iris, hue='species', diag_kind='hist');
```

```python
plt.subplots(figsize = (8,6))
sns.boxplot(data = iris, x = 'species', y = 'sepal_length');
```
