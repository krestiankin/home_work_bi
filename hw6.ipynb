>6. Покажите подробную информацию о наименее и наиболее прибыльных фильмах в наборе данных.

```python
df['Worldwide Gross'] = df['Worldwide Gross'].str.strip('$').astype(float) # прибираємо знак $ та переводино значення у флоат

df_wwg_min = df['Worldwide Gross'].min()       # роблю через змінні щоб не був довгий рядок коду
df_wwg_max = df['Worldwide Gross'].max()       

df[(df['Worldwide Gross'] == df_wwg_min)|(df['Worldwide Gross'] == df_wwg_max)] # виводимо по масці 2 рядки з мін та макс знач
```

ще рішення по колонці Profitability, напевно більш правильне

```python
df_wwg_min = df['Profitability'].min()       # роблю через змінні щоб не був довгий рядок коду
df_wwg_max = df['Profitability'].max()       

df[(df['Profitability'] == df_wwg_min)|(df['Profitability'] == df_wwg_max)] # виводимо по масці рядки з мін та макс значеннями
```

> 7. Значение «Жанр» временами кажется непоследовательным; попробуйте найти эти несоответствия и исправить их.

```python
df.Genre.value_counts() # бачимо що унікальні назви трохи написані з помилками 'Romence', 'romance', 'Comdy', 'comedy', робимо замуну на вірні.
```

```python
df['Genre'] = df.Genre.replace(['Romence', 'romance', 'Comdy', 'comedy'], ['Romance', 'Romance', 'Comedy', 'Comedy'])
df.Genre.value_counts()   # робимо знову перевірку, тепер все норм.
```

```python
df[df.duplicated()] #перевіряємо наявність дублікатів
```

```python
df.drop_duplicates(inplace=True) # видаляємо дублікати
df[df.duplicated()] # перевіряємо чи пройшла операція видалення дублікатів
```
