```python
adress = './orders.txt'
```

```python
with open(adress, 'r') as f:
    data = f.read().split('\n\n')
    
# генерируем лист из ордеров
orders = [order.split('@@@') for order in data]

#print(orders)
print('Total orders: ', len(orders))
print(len(orders[0]))
```

```python
orders[0:3]
```

```python
from collections import defaultdict 
import sys
import os
```

```python
product_count = defaultdict(lambda: 0)
product_pair_counts = defaultdict(int)

for order in orders:
    for p in order:
        # добавление в словарь уникальные частоты товаров в чекаъ
        product_count[p] += 1
    for i, p1 in enumerate(order[::-1]):
        for p2 in order[i + 1:]:
            product_pair_counts[(p1, p2)] += 1
            product_pair_counts[(p2, p1)] += 1

print(len(product_count))                            
```

```python
dict(product_count)
```

```python
# создаем генератор словаря с расчетов доверия, count - частота встречаемости

confidence = {
    (p1, p2): count / product_count[p1] for (p1, p2),
    count in product_pair_counts.items()
}
```

```python
confidence
```

```python
def print_asscotiations(min_conf, min_support, orders_list):
    len_orders_list = len(orders_list)
    for (p1, p2), count in product_pair_counts.items():
        support = count / len_orders_list 
        conf = confidence[(p1, p2)]
        if conf >= min_conf and support >= min_support:
            print(f'{p1} => {p2} ({conf * 100} confidence), {support * 100} support')
```

```python
print_asscotiations(.20, 0.01, orders)
```

```python
# Next varient to solve
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(orders)

df = pd.DataFrame(te_ary, columns= te.columns_)
df.head(7)
```

```python
frequent_itemsets = apriori(df, min_support= 0.01, max_len= 2, use_colnames = True)
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold= 0.2)
```
